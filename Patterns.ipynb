```python
size = 5
for i in range(size):
    if i == 0 or i == size - 1:
        print("*" * size)
    else:
        print("*" + " " * (size - 2) + "*")
```

```python
rows = 4
cols = 8
for i in range(rows):
    if i == 0 or i == rows - 1:
        print("#" * cols)
    else:
        print("#" + " " * (cols - 2) + "#")
```
