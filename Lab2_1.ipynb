```python
import gradio as gr

timeline_text = """
### 📅 Computer History Timeline

**3000 BCE** — Abacus was used for counting.
**1642** — Blaise Pascal invented Pascaline.
**1837** — Charles Babbage designed the Analytical Engine.
**1940s** — First generation computers using vacuum tubes appeared.
**1950s** — Transistor-based computers were introduced.
**1970s** — Microprocessors changed the computer industry.
**1990s** — Internet became popular.
**Present** — Artificial Intelligence and quantum computing are growing rapidly.
"""

inventors_text = """
### 👩‍💻 Famous Inventors and Contributors

**Charles Babbage** — Known as the Father of the Computer.

**Ada Lovelace** — Considered the first computer programmer.

**Alan Turing** — Contributed to computer science and artificial intelligence.

**John von Neumann** — Developed the stored-program computer architecture.

**Tim Berners-Lee** — Invented the World Wide Web.
"""

generation_data = {
    "First Generation": """
### First Generation

**Period:** 1940–1956
**Main Technology:** Vacuum Tubes
**Examples:** ENIAC, UNIVAC
**Features:** Large size, high heat, expensive, slow processing
""",
    "Second Generation": """
### Second Generation

**Period:** 1956–1963
**Main Technology:** Transistors
**Examples:** IBM 1401, IBM 7094
**Features:** Smaller, faster, more reliable than first generation
""",
    "Third Generation": """
### Third Generation

**Period:** 1964–1971
**Main Technology:** Integrated Circuits
**Examples:** IBM System/360
**Features:** Better speed, smaller size, lower cost
""",
    "Fourth Generation": """
### Fourth Generation

**Period:** 1971–Present
**Main Technology:** Microprocessors
**Examples:** Personal computers, laptops
**Features:** Portable, affordable, powerful
""",
    "Fifth Generation": """
### Fifth Generation

**Period:** Present and Future
**Main Technology:** Artificial Intelligence
**Examples:** AI systems, robots, quantum computers
**Features:** Machine learning, natural language processing, automation
"""
}

def show_generation(generation):
    return generation_data[generation]

def quiz(q1, q2, q3):
    score = 0

    if q1 == "Vacuum Tubes":
        score += 1
    if q2 == "Charles Babbage":
        score += 1
    if q3 == "Fourth":
        score += 1

    if score == 3:
        return f"Your score is {score}/3 🎉 Excellent!"
    elif score == 2:
        return f"Your score is {score}/3 👍 Good work. Revise one more time."
    else:
        return f"Your score is {score}/3 📘 Keep learning and try again."

with gr.Blocks(title="History of Computers") as app:
    gr.Markdown("# 💻 History of Computers")
    gr.Markdown("Python and Gradio.")

    with gr.Tab("Home"):
        gr.Markdown("""
        ## Welcome

        This app explains the history of computers in a simple and interactive way.

        Students can learn about:

        - Early computing devices
        - Five generations of computers
        - Famous inventors
        - Important milestones
        """)

    with gr.Tab("Timeline"):
        gr.Markdown(timeline_text)

    with gr.Tab("Generations"):
        gen_dropdown = gr.Dropdown(
            choices=list(generation_data.keys()),
            value="First Generation",
            label="Select a generation"
        )
        gen_output = gr.Markdown()
        gen_dropdown.change(show_generation, inputs=gen_dropdown, outputs=gen_output)
        app.load(show_generation, inputs=gen_dropdown, outputs=gen_output)

    with gr.Tab("Famous Inventors"):
        gr.Markdown(inventors_text)

    with gr.Tab("Quiz"):
        q1 = gr.Radio(
            ["Transistors", "Vacuum Tubes", "Microprocessors", "Artificial Intelligence"],
            label="1. Which technology was used in first generation computers?"
        )

        q2 = gr.Radio(
            ["Alan Turing", "Charles Babbage", "Tim Berners-Lee", "Bill Gates"],
            label="2. Who is known as the Father of the Computer?"
        )

        q3 = gr.Radio(
            ["First", "Second", "Third", "Fourth"],
            label="3. Which generation used microprocessors?"
        )

        submit = gr.Button("Submit Quiz")
        result = gr.Textbox(label="Result")

        submit.click(quiz, inputs=[q1, q2, q3], outputs=result)

app.launch(share=True)
```
